```python
import matplotlib.pyplot as plt   # pip install matplotlib
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels = [] # NULL LIST OF PYTHON
file_name = 'Labels.txt'
with open(file_name,'rt') as fpt:
     classLabels = fpt.read().rstrip('\n').split('\n')
     #classLabels.append(fpt.read())                                            
                                              
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)   ## 255/2 = 127.5
model.setInputMean((127.5,127.5,127.5))  # mobilenet takes input as =>  [-1,1]
model.setInputSwapRB(True)
```

# read an image

```python
img = cv2.imread('car6.png')
```

```python
plt.imshow(img)
```

```python
plt.imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidence ,bbox = model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    #cv2.rectangle(frame, (x,y),(x+w, y+h), (255, 0, 0), 2)
    #cv2.pvtText(img, text, (text_offset_x, test_offest_y), font, fontScalefont_scale, color=(0, 0, 0), thickness=1)
    cv2.rectangle(img,boxes,(255, 0, 0), 2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale,color=(0,255,0), thickness=3)
    
```

# Accessing object in a video

```python
import cv2

cap = cv2.VideoCapture('video.mp4')

# Check if the video is opened correctly 
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open video")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    print(ClassIndex)
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:
                cv2.rectangle(frame, (boxes[0], boxes[1]), (boxes[2], boxes[3]), (255, 0, 0), 2)
                cv2.putText(frame, classLabels[ClassInd - 1], (boxes[0] + 40, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

    cv2.imshow('Object Detection', frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
